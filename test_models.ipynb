```python
#Import the model from your implementation
from state_space_models import ETS_AAdM
#Import the corresponding model from statsmodels
from statsmodels.tsa.holtwinters import ExponentialSmoothing


#Create a function to initialize both models
#Make sure the return order is (Implemented, Stasmodels)
#If you want to make the function take more arguments then just data;
#Go to the test.py in testing directory and under the .test_model() method change the provided arguments to initialize_models function that it takes as an argument
#The function is called in line 86 in test.py

def initialize_models(data):

    model1 = ETS_AAdM(dep_var=data, seasonal_periods=12)
    model2 = ExponentialSmoothing(data, trend="add", damped_trend="true", seasonal="mul", seasonal_periods=12, initialization_method="heuristic")
    
    return model1, model2
```
